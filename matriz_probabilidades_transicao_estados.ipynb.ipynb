```python
def get_gene_truth_table (gene):
    truth_table_gene =  truth_table[truth_table['GENE'] == gene]
    truth_table_gene = truth_table_gene.dropna(axis='columns') 
    return truth_table_gene
```

```python
tt_gene = get_gene_truth_table('X1')
tt_gene
```

```python
#gene: gene
#value: expected value for gene
#x1 , x2 , x3 , x4: actual values
def get_gene_probability(gene , value , x1 , x2 , x3 , x4 ):
    tt_gene = get_gene_truth_table(gene)
    search = []
    for dependency in gene_dependencies[gene]:
        if dependency == 'X1':
            search.append(x1) 
        elif dependency == 'X2':
            search.append(x2) 
        elif dependency == 'X3':
            search.append(x3) 
        elif dependency == 'X4':
            search.append(x4)
            
    tt_gene = tt_gene.loc[(tt_gene['INDEX'] == ''.join(search))]
    
    if value == '0':
        return tt_gene.iloc[0]['P0']
    else:
        return tt_gene.iloc[0]['P1']
```

```python
#P( state_to | state_from) => P( 0000 | 0000 ) = P( X1 = 0 | 0000) * P( X2 = 0 | 0000) * P( X3 = 0 | 0000) * P( X4 = 0 | 0000)  =  1*1 *1*1 = 1
def get_transition_probability_from_state1_to_state2 (state_from, state_to):
    #print(state_from + " : " +state_to)
    x1 = state_from[0]
    x2 = state_from[1]
    x3 = state_from[2]
    x4 = state_from[3]
    
    x1_prob = get_gene_probability('X1' , state_to[0] , x1, x2 , x3 , x4 )
    x2_prob = get_gene_probability('X2' , state_to[1] , x1, x2 , x3 , x4 )
    x3_prob = get_gene_probability('X3' , state_to[2] , x1, x2 , x3 , x4 )
    x4_prob = get_gene_probability('X4' , state_to[3] , x1, x2 , x3 , x4 )
    
   # print("state2[0] " + state2[0])
   # print("x1_prob " + str(x1_prob))
   # print("x2_prob " + str(x2_prob))
   # print("x3_prob " + str(x3_prob))
   # print("x4_prob " + str(x4_prob))

    return x1_prob * x2_prob * x3_prob * x4_prob
```

```python
for state_from in states:
    for state_to in states:
        probability = get_transition_probability_from_state1_to_state2(state_from, state_to)
        #print('probability: ' + str(probability ) )
        probability_matrix.xs(state_from)[state_to] = probability
```
